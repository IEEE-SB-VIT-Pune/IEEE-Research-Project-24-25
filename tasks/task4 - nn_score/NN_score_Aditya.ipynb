```python
import numpy as np
import random
from sklearn.neighbors import KNeighborsClassifier
```

```python
def create_matrix(row, column):
    return np.random.randint(1, 21, size=(row, column), dtype=np.int32)
```

```python
def print_matrix(matrix):
    for row in matrix:
        print(' '.join(map(str, row)))
```

```python
m=int(input("Enter number of rows:"))
n=int(input("Enter number of columns:"))
```

```python
matrix1=create_matrix(m,n)
print_matrix(matrix1)
```

```python

def calc_nn(matrix,n_neighbors):
    knn = KNeighborsClassifier(n_neighbors=n_neighbors+1,metric='manhattan')
    label=[x for x in range(len(matrix))]
    knn.fit(matrix,label)
    indices=knn.kneighbors(matrix,return_distance=False)
    indices=indices[:, 1:]
    nn_score = np.zeros(matrix.shape[0], dtype=int)
    for x in indices:
        for y in x:
            nn_score[y] += 1
    return nn_score
```

```python
nn_scorelist = calc_nn(matrix1, n_neighbors=3)

for i in range(len(nn_scorelist)):
    print("NN Score for row"+str(i+1)+":"+str(nn_scorelist[i]))
```
