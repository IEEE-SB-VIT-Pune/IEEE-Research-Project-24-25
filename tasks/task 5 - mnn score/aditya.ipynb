```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
def create_matrix(row, column):
    return np.random.randint(1, 21, size=(row, column), dtype=np.int32)
```

```python
def print_matrix(matrix):
    for row in matrix:
        print(' '.join(map(str, row)))
```

```python
m=int(input("Enter number of rows:"))
n=int(input("Enter number of columns:"))
```

```python
matrix1=create_matrix(m,n)
print_matrix(matrix1)
```

```python
def precompute(matrix, n_neighbors):
    knn = KNeighborsClassifier(n_neighbors=n_neighbors+1, metric="manhattan")
    label = [x for x in range(len(matrix))]
    knn.fit(matrix, label)
    indices = knn.kneighbors(matrix, return_distance=False)
    indices = indices[:, 1:]
    neighbor_dict = {}
    for i, neighbors in enumerate(indices):
        neighbor_dict[str(i)] = set(neighbors)
    return neighbor_dict
```

```python
data_1=precompute(matrix1,4)
print(data_1)
```

```python

def calc_mnn(matrix, n_neighbors, data):
    knn = KNeighborsClassifier(n_neighbors=n_neighbors+1, metric='manhattan')
    labels = [x for x in range(len(matrix))]
    knn.fit(matrix, labels)
    indices = knn.kneighbors(matrix, return_distance=False)
    indices = indices[:, 1:]  # Exclude the point itself

    mnn_score = np.zeros(matrix.shape[0], dtype=int)

    for i, neighbors in enumerate(indices):
        for neighbor in neighbors:
            if i in data.get(str(neighbor), set()):
                mnn_score[i] += 1
    return mnn_score
```

```python
score=calc_mnn(matrix1,4,data_1)
print(score)
```
